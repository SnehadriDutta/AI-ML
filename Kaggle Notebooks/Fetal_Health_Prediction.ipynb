# Model Building

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
X = data.drop('fetal_health', axis = 1)
Y = data['fetal_health']
```

```python
sm = SMOTE(random_state = 42)
X_resem, Y_resem = sm.fit_resample(X,Y)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X_resem, Y_resem, test_size = 0.2, random_state = 0)
```

### LGBMClassifier

```python
# LGBMClassifier
lgbm = LGBMClassifier()
lgbm.fit(xtrain, ytrain)
print(f'Model Score {lgbm.score(xtest,ytest)}')
```

```python
lgbm_pred = lgbm.predict(xtest)
print('Confusion Matrix')
matrix = confusion_matrix(ytest,lgbm_pred)
sns.heatmap(matrix, annot = True)
```

```python
print('Classification Report')
print(classification_report(ytest, lgbm_pred))
```

### AdaBoosting

```python
rdf = RandomForestClassifier()
adaboost = AdaBoostClassifier(base_estimator = rdf, n_estimators = 200, random_state = 0)
adaboost.fit(xtrain, ytrain)
adaboost_pred = adaboost.predict(xtest)
print(f'Model Score: {adaboost.score(xtest, ytest)}')
```

```python
print('Confusion Matrix')
matrix = confusion_matrix(ytest,adaboost_pred)
sns.heatmap(matrix, annot = True)
```

```python
print('Classification Report')
print(classification_report(ytest, adaboost_pred))
```

## Neural Networks

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, Flatten, MaxPooling1D
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))
xtrain_norm = scaler.fit_transform(xtrain)
xtest_norm = scaler.transform(xtest)
```

```python
ytrain_test = ytrain - 1
ytrain_test
```

```python
from tensorflow.keras.utils import to_categorical

ytrain_new = ytrain - 1 # to convert the labels to 0-2
ytest_new = ytest - 1
ytrain_cat = to_categorical( ytrain_new, num_classes=3)
ytest_cat = to_categorical(ytest_new , num_classes=3)
```

### Simple ANN

```python
ann = Sequential()
ann.add(Dense(64, input_dim = (xtrain.shape[1]), activation = 'relu'))
ann.add(Dense(124, activation = 'relu'))
ann.add(Dense(3, activation = 'softmax'))
ann.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
ann.summary()
```

```python
ann.fit(xtrain_norm, ytrain_cat, epochs = 100)
```

```python
ann_pred = np.argmax(ann.predict(xtest_norm), axis=1)
print('Confusion Matrix')
matrix = confusion_matrix(ytest_new, ann_pred)
sns.heatmap(matrix, annot = True)
```

```python
print('Classification Report')
print(classification_report(ytest_new, ann_pred))
```

### Convolution Model

```python
xtrain_cnn = xtrain_norm.reshape(xtrain_norm.shape[0], xtrain_norm.shape[1], 1)
xtest_cnn = xtest_norm.reshape(xtest_norm.shape[0], xtest_norm.shape[1], 1)
```

```python
cnn = Sequential()
cnn.add(Conv1D(64, 2, input_shape = (xtrain_cnn.shape[1], 1), activation = 'relu'))
cnn.add(Dense(124, activation = 'relu'))
cnn.add(Flatten())
cnn.add(Dense(3, activation = 'softmax'))
cnn.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
cnn.summary()
```

```python
cnn.fit(xtrain_cnn, ytrain_cat, epochs = 150)
```

```python
cnn_pred = np.argmax(cnn.predict(xtest_norm), axis=1)
print('Confusion Matrix')
matrix = confusion_matrix(ytest_new, cnn_pred)
sns.heatmap(matrix, annot = True)
```

```python
print('Classification Report')
print(classification_report(ytest_new, cnn_pred))
```
