```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("C:\\Users\\hp\\Desktop\\Misc\\Programming\\Machine & Deep Learning\\Deep Learning\\Sample Data\\Stock-Sentiment-Analysis-master\\Data.csv")
data.head()
```

```python
traindata = data[data.Date < '20150101']
testdata = data[data.Date > "20150101"]
```

```python
traindata
```

```python
# Removing Punctuation
xtrain = traindata.iloc[:,2:27]
xtrain.replace("[^a-zA-Z]", " ", regex = True, inplace = True)
xtrain
```

```python
# Changing the headers

list_ = [str(i) for i in range(0,25)]
xtrain.columns = list_
xtrain.head()
```

```python
# Coverting all the text into lower case
for i in range(0,25):
    xtrain[str(i)] = xtrain[str(i)].str.lower()
xtrain.head()
```

```python
# Joining all the columns 
headlines = []
rows = len(xtrain.index)
for row in range(0,rows):
    headlines.append(" ".join(str(x) for x in xtrain.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier

countvec = CountVectorizer(ngram_range = (2,2))
xtraindata = countvec.fit_transform(headlines)
```

```python
model = RandomForestClassifier()
model.fit(xtraindata,traindata.Label)
```

```python
xtest = []
rows = len(testdata.index)
for row in range(0,rows):
    xtest.append(" ".join(str(x) for x in testdata.iloc[row,2:25]))
```

```python
import re
for i in range(len(xtest)):
    
    xtest[i] = xtest[i].lower()
    xtest[i] = re.sub("[^a-zA-Z]", " ",xtest[i])
xtest[0]
```

```python
xtest = countvec.transform(xtest)
ypred = model.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print(accuracy_score(testdata.Label, ypred))
```
