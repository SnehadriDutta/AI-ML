```python
import pandas as pd
import numpy as np
from tensorflow import keras
```

```python
data = pd.read_csv("insurance.csv")
data.shape
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data.sex = le.fit_transform(data.sex)
data.smoker = le.fit_transform(data.smoker)
```

```python
data.smoker.value_counts()
```

```python
data.head()
```

```python
data = pd.get_dummies(data)
```

```python
data.head()
```

```python
X = data.drop('expenses', axis=1).values
Y = data['expenses'].values
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X,Y, test_size=0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

Building ANN

```python
ann = keras.models.Sequential()
ann.add(keras.layers.Dense(units=14, kernel_initializer='he_uniform', activation='relu'))
```

```python
ann.add(keras.layers.Dense(units=14, kernel_initializer='he_uniform', activation='relu'))
```

```python
ann.add(keras.layers.Dense(units=1, kernel_initializer='he_uniform', activation='relu'))
```

```python
ann.compile(optimizer='adam', loss='mean_absolute_error', metrics=['mean_absolute_error'])
```

```python
ann.fit(xtrain,ytrain, batch_size =32, epochs=100)
```
